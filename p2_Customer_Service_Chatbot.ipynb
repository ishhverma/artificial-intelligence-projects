```python
import nltk
import string
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

# Download necessary NLTK data
nltk.download('wordnet', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('punkt_tab', quiet=True) # Added to fix LookupError: Resource punkt_tab not found.
```

```python
synthetic_faq_data = [
    {
        "question": "How do I check my order status?",
        "answer": "You can check your order status by logging into your account and visiting the 'Order History' page, or by entering your order number on our 'Track Order' page."
    },
    {
        "question": "What is your return policy?",
        "answer": "Our return policy allows for returns within 30 days of purchase, provided the item is unused and in its original packaging. Please see our 'Returns' page for full details."
    },
    {
        "question": "How can I contact customer support?",
        "answer": "You can contact our customer support team via email at support@example.com, by calling us at 1-800-123-4567, or through our live chat feature on the website."
    },
    {
        "question": "Can I change my shipping address after placing an order?",
        "answer": "Unfortunately, once an order is placed, we cannot guarantee changes to the shipping address. Please contact support immediately if you need to request a change."
    },
    {
        "question": "Do you offer international shipping?",
        "answer": "Yes, we offer international shipping to most countries. Shipping costs and delivery times vary by destination. You can see the options at checkout."
    }
]

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = nltk.word_tokenize(text)
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(words)

cleaned_questions = []
cleaned_answers = []

for item in synthetic_faq_data:
    cleaned_questions.append(preprocess_text(item['question']))
    cleaned_answers.append(item['answer'])

questions = cleaned_questions
answers = cleaned_answers

print("Data preprocessing complete. `questions` and `answers` lists are ready.")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Instantiate TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the questions into TF-IDF vectors
question_vectors = tfidf_vectorizer.fit_transform(questions)

def recognize_intent(user_query):
    # Preprocess the user query
    preprocessed_query = preprocess_text(user_query)

    # Transform the preprocessed query into a TF-IDF vector using the fitted vectorizer
    query_vector = tfidf_vectorizer.transform([preprocessed_query])

    # Calculate cosine similarity between the query vector and all question vectors
    similarities = cosine_similarity(query_vector, question_vectors)

    # Get the index of the question with the highest similarity score
    most_similar_index = similarities.argmax()
    highest_similarity_score = similarities[0, most_similar_index]

    return most_similar_index, highest_similarity_score

print("TF-IDF vectorization and intent recognition function `recognize_intent` are set up.")
```

```python
import re

def extract_entities(user_query):
    entities = {}
    # Example: Extracting a potential order ID (a sequence of digits)
    order_id_match = re.search(r'\b(\d{5,})\b', user_query) # Matches 5 or more digits as a word boundary
    if order_id_match:
        entities['order_id'] = order_id_match.group(1)

    # Add more entity extraction logic here if needed

    return entities

print("Entity extraction function `extract_entities` is set up.")
```

```python
order_status_intent_index = questions.index('check order status')
similarity_threshold = 0.45 # Using the adjusted threshold

def chatbot_response(user_query):
    # Extract entities first
    entities = extract_entities(user_query)

    most_similar_index, highest_similarity_score = recognize_intent(user_query)

    response = ""
    if highest_similarity_score >= similarity_threshold:
        # Check for 'check order status' intent with missing order_id
        if most_similar_index == order_status_intent_index and 'order_id' not in entities:
            response = "I can help you with your order status. Could you please provide your order number?"
        # Check for specific entity-aware responses (e.g., order status with order ID)
        elif 'order_id' in entities and most_similar_index == order_status_intent_index:
            order_id = entities['order_id']
            response = f"You mentioned order {order_id}. You can check its status by logging into your account and visiting the 'Order History' page, or by entering your order number on our 'Track Order' page."
        else:
            # General FAQ answer if no specific entity logic applies or entity not relevant
            response = answers[most_similar_index]
    else:
        response = "I'm sorry, I don't understand your question. Could you please rephrase it or ask something different?"

    return response

print("Chatbot response function `chatbot_response` is consolidated and ready.")
```

```python
print("\n--- Chatbot Interactive Session ---")
print("Type 'exit', 'quit', or 'bye' to end the conversation.")

while True:
    user_input = input('You: ')
    if user_input.lower() in ['exit', 'quit', 'bye']:
        print('Chatbot: Goodbye!')
        break

    response = chatbot_response(user_input)
    print(f'Chatbot: {response}')
```
